##Chi-square Test

```python
from scipy.stats import chi2_contingency
```

```python
# Create a contingency table of 'type' and 'country'
contingency_table_country = pd.crosstab(data['type'], data['country'])

# Perform the Chi-Squared test
chi2_country, p_country, dof_country, expected_country = chi2_contingency(contingency_table_country)

print(f"Chi-Squared statistic for Type and Country: {chi2_country}")
print(f"P-value for Type and Country: {p_country}")
print(f"Degrees of freedom for Type and Country: {dof_country}")
print("Expected frequencies table for Type and Country:")
print(expected_country[0])
```

```python
# Create a contingency table of 'type' and 'rating'
contingency_table = pd.crosstab(data['type'], data['rating'])

# Perform the Chi-Squared test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-Squared statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies table:")
print(expected)
```

The results show a very small which is much less than the typical significance level of 0.05. This indicates that there is a statistically significant relationship between the type of content (Movie or TV Show) and its rating. In other words, the distribution of ratings is significantly different for movies and TV shows on Netflix.

Text Preprocessing: In order to remove stop words and create better wordclouds in disambiguation

```python
import nltk
import re
from nltk.corpus import stopwords

# Download stopwords if not already present
try:
    stopwords.words('english')
except LookupError:
    nltk.download('stopwords')

stops = set(stopwords.words("english"))

def text_preprocessing(columns_pandas,stop_words):
    # Use regex to find all alphabetic words
    tokens = re.findall(r'\b[a-zA-Z]+\b', columns_pandas)
    tokens = [word for word in tokens if word.lower() not in stop_words]
    preprocessed_text = " ".join(tokens)
    return preprocessed_text

# Use the 'title' column for text preprocessing
data['Preprocessed_Description'] = data.apply(lambda x: text_preprocessing(x['title'],stops), axis=1)
```

#Recommendation Engine

## Content-Based Filtering

Content-based filtering recommends items to a user based on the characteristics of items the user has previously liked. In this case, we will use the text descriptions and genres of the movies and TV shows to find similar items.

We will use the TF-IDF (Term Frequency-Inverse Document Frequency) vectorizer to convert the text data into a numerical representation. Then, we will calculate the cosine similarity between items to find those that are most similar.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Combine relevant features into a single string
data['combined_features'] = data['title'] + ' ' + data['director'] + ' ' + data['country'] + ' ' + data['genre'] + ' ' + data['rating']

# Initialize the TF-IDF vectorizer
tfidf_vectorizer = TfidfVectorizer(stop_words='english')

# Fit and transform the combined features
tfidf_matrix = tfidf_vectorizer.fit_transform(data['combined_features'])

# Calculate the cosine similarity matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

Now that we have the cosine similarity matrix, we can create a function to get recommendations based on the similarity scores.

```python
# Create a mapping of titles to indices
indices = pd.Series(data.index, index=data['title']).drop_duplicates()

def get_recommendations(title, cosine_sim=cosine_sim, data=data, indices=indices):
    # Get the index of the title that matches the input title
    idx = indices[title]

    # Get the pairwise similarity scores for all titles with that title
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the titles based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar titles (excluding the input title itself)
    sim_scores = sim_scores[1:11]

    # Get the title indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar titles
    return data['title'].iloc[movie_indices]
```

Let's test the recommendation system with an example.

```python
# Get recommendations for a specific title
recommendations = get_recommendations('Sankofa')

# Display the recommendations
print("Recommendations for 'Sankofa':")
display(recommendations)
```

The Content-Based Filtering approach provides recommendations based on the characteristics of the content itself (title, director, genre, etc.), which is useful when user interaction data is limited.
